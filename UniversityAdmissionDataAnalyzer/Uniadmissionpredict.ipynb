```python
#apply a backward elimination process. 
#we change data type to float as well since its an object
#create a new feature vector x_opt, which will only contain a set of independent features that are significantly affecting the dependent variable.
#Next we need to choose a significant level(0.5), and then need to fit the model with all possible predictors. 
#So for fitting the model, we will create a regressor_OLS object of new class OLS of statsmodels library. 
#Then we will fit it by using the fit() method.
#p-value to compare with SL value, so we use summary() method to get the summary table.
x_opt = z[:,[0,1,2,3,4,5,6]]
x_opt = np.array(x_opt, dtype = float)
regressor_OLS = sm.OLS(endog = w, exog = x_opt).fit()
regressor_OLS.summary()
```

```python
#Now OLS stands for ‘Ordinary Least Squares’, which essentially trains a Linear model. 
#we observe that the higest P-value is for feature 4, which is way over our significance level of 0.05. 
#Hence, we remove it.
x_opt = z[:,[0,1,2,4,5,6]]
x_opt = np.array(x_opt, dtype = float)
regressor_OLS = sm.OLS(endog = w, exog = x_opt).fit()
regressor_OLS.summary()
```
